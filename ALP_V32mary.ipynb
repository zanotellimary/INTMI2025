##Operadores

adição+

subtração-

divisão/

multiplicação*

resto da divisão %

divisão inteira//

exponenciação**

```python
numero1 = 10
numero2 = 5

print("a soma é igual a: ", numero1 + numero2)
print("a subtração é igual a: ", numero1 - numero2)
print("a divisão é igual a: ", numero1 / numero2)
print("a multiplicaçõa é igual a: ", numero1 * numero2)
print("a resto é igual a: ", numero1 % numero2)
print("a divisão inteira é igual a: ", numero1 // numero2)
print("a exponencial é igual a: ", numero1 ** numero2)
```
